_ I'm excited to try plotting my own figures using standard scientific python libraries _

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from IPython.html.widgets import interact

# x = np.linspace(0,10,1000)
# plt.plot(x,np.sin(x))
# plt.plot(x,np.sin(x**2))

def plot_sine(frequency=1.0,amplitude=1.0):
    plt.ylim(-1.0,1.0)
    x = np.linspace(-10,10,1000)
    plt.plot(x, amplitude*np.cos(frequency*(x**2)))
    
interact(plot_sine, frequency=(0.5,10.0), amplitude=(0.1,2.0))
```
